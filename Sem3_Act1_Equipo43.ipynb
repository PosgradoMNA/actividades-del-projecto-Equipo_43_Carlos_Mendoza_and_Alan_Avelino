##**Información del conjunto de datos:**

Esta investigación apuntó al caso de los pagos predeterminados de los clientes en Taiwán y compara la precisión predictiva de la probabilidad de incumplimiento entre seis métodos de minería de datos. Desde la perspectiva de la gestión de riesgos, el resultado de la precisión predictiva de la probabilidad de incumplimiento estimada será más valioso que el resultado binario de la clasificación: clientes creíbles o no creíbles. Debido a que se desconoce la probabilidad real de incumplimiento, este estudio presentó el novedoso "Método de suavizado de clasificación" para estimar la probabilidad real de incumplimiento. Con la probabilidad real de incumplimiento como variable de respuesta (Y) y la probabilidad predictiva de incumplimiento como variable independiente (X), el resultado de la regresión lineal simple (Y = A + BX) muestra que el modelo de pronóstico producido por la red neuronal artificial tiene el coeficiente de determinación más alto; su intersección de regresión (A) es cercana a cero y el coeficiente de regresión (B) a uno. Por lo tanto, entre las seis técnicas de minería de datos, la red neuronal artificial es la única que puede estimar con precisión la probabilidad real de incumplimiento.

```python
#Importando la librería pandas
import pandas as pd
```

Obtengo la información de mi Base de Datos, se la asigno a una variable y defino que su index empiece en 1. Finalmente se utiliza "describe" para visualizar el detalle de cada columna.

```python
df = pd.read_csv("default of credit card clients.csv", index_col=0)
df.describe()
```

Para empezar la limpieza de los datos, se identifica en qué columnas se enceuntran los valores Nulos y valores NaN.

```python
df.isnull().any()
```

```python
df.isna().any()
```

Se realiza una copia de la base original para dar formato a las columnas.

```python
df_2 = df.copy()
df_2.rename(columns= {'X1':'MONTO_OTORGADO','X2':'GENERO','X3':'EDUCACION','X4':'ESTADO_CIVIL','X5':'EDAD','X6':'PAGO_SEP_2005','X7':'PAGO_AGO_2005','X8':'PAGO_JUL_2005','X9':'PAGO_JUN_2005','X10':'PAGO_MAY_2005','X11':'PAGO_ABR_2005'}, inplace=True)
df_2.rename(columns= {'X12':'EDO_CUE_SEP_2005','X13':'EDO_CUE_AGO_2005','X14':'EDO_CUE_JUL_2005','X15':'EDO_CUE_JUN_2005','X16':'EDO_CUE_MAY_2005','X17':'EDO_CUE_ABR_2005'}, inplace=True)
df_2.rename(columns= {'X18':'MON_PAG_SEP_2005','X19':'MON_PAG_AGO_2005','X20':'MON_PAG_JUL_2005','X21':'MON_PAG_JUN_2005','X22':'MON_PAG_MAY_2005','X23':'MON_PAG_ABR_2005','Y':'VAR_RESPUESTA'}, inplace=True)
df_2
```

Del resultado anterior se eliminan los NaN que pudieran existir en cada columna. Se ejecuta el describe para ver resultado de cada columna.

```python
df_2.dropna(inplace=True)
df_2.describe()
```

Se obtienen los valores máximos y mínimos para identificar si se puede realizar alguna normalización de los datos.

```python
df_2.min()
```

```python
df_2.max()
```

La columna GENERO cumple con lo establecido en la definición del DataSet. Procedemos con la columna EDUCACION, donde la muestra nos dice que hay valores menores a 1 y mayores a 4. Por tanto la desición que se tomó fue modificar esos valores a 4 que sería "Otros".

```python
df_2.loc[df_2["EDUCACION"]>4,"EDUCACION"]=4
df_2.loc[df_2["EDUCACION"]<1,"EDUCACION"]=4
```

Se vuelven a generar las validaciones de valores mínimo y máximo para visualizar si se corrigió.

```python
df_2.min()
```

```python
df_2.max()
```

Visualizando los datos mínimos y máximos para la columna ESTADO_CIVIL, aplicamos el mismo criterio que con EDUCACION, el cambio sería que el valor 3 = otros. Ya aplicando esto, validamos de nuevo los valores.

```python
df_2.loc[df_2["ESTADO_CIVIL"]>3,"ESTADO_CIVIL"]=3
df_2.loc[df_2["ESTADO_CIVIL"]<1,"ESTADO_CIVIL"]=3
```

```python
df_2.min()
```

```python
df_2.max()
```

Con base en la definición del DataSet, se menciona que las columnas X6 a X11 (PAGO_SEP_2005-PAGO_ABR_2005) los valores van de -1 a 9 con su significado. Dado lo anterior, se analizaron los datos contenidos en el DataSet van de -2 a 8. 
Por lo tanto, se tomó la desición de eliminar estás columnas puesto que no tenemos una regla que nos indique el valor de los -2 y 0, tomando futuramente como base de las columnas X12 a X23.

```python
df_3 = df_2.drop(['PAGO_SEP_2005', 'PAGO_AGO_2005', 'PAGO_JUL_2005', 'PAGO_JUN_2005', 'PAGO_MAY_2005', 'PAGO_ABR_2005'], axis = 1)
df_3.head(19)
```
